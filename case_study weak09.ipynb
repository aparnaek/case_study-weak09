# 1.The mean sales generated by each zone

```python
data_mean=data[['Zone - A','Zone - B','Zone - C','Zone - D']].mean()
print(data_mean)
```

# 2.Total sales generated by all the zones for each month

# 3.Checking whether all the zones generate the same amount of sales

### H0:-amount of sales in all zone are equal

### H1:-amount of sales in each zone differ

```python
from scipy.stats import f_oneway
f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

by applying anova test we get f value as 5.672 and p value 0.001 that is nearly equal to 0.so that we reject the null hypothesis and accept alternative hypothesis that is amount of sales in each zone is different
